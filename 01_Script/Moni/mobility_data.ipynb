### Import Libraries

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

This notebook will handle all the mobility data, like creating new csv files with smaller aspects of the data.

```python
# read in the raw data again
df_raw = pd.read_csv("../../02_Data//05_Mobility/2020_US_Region_Mobility_Report.csv")
```

```python
# take a quick look
df_raw.head()
```

```python
# get all the national data and separate to its own csv file
df_national = df_raw.loc[df_raw['sub_region_1'].isnull()]
```

```python
# drop preliminary unnecessary columns
df_national = df_national.drop(columns=['country_region_code', 'country_region'])
```

```python
# the two sub_region_columns are also unnecessary now
df_national = df_national.drop(columns=['sub_region_1', 'sub_region_2'])
```

```python
# finally, the two sub_region_columns are also unnecessary now
df_national = df_national.drop(columns=['metro_area', 'iso_3166_2_code', 'census_fips_code'])
```

```python
df_national.head()
```

```python
df_national.columns
```

```python
# send new df to its own csv file
df_national.to_csv(path_or_buf="../../02_Data//05_Mobility/US_Mobility_Report.csv", columns=df_national.columns, index=False)
```

Now let's do the same for all the states. We will save the data for all the states in its own csv file.

```python
df_raw.head()
```

```python
df_states = df_raw.loc[df_raw['sub_region_1'].notnull() & df_raw['sub_region_2'].isnull()]
```

```python
# again, drop all columns that aren't necessary
df_states = df_states.drop(columns=['country_region_code', 'country_region', 'sub_region_2', 'metro_area', 'iso_3166_2_code', 'census_fips_code'])
```

```python
df_states.describe()
```

```python
# send new df to its own csv file
df_states.to_csv(path_or_buf="../../02_Data//05_Mobility/US_States_Mobility_Report.csv", columns=df_states.columns, index=False)
```
